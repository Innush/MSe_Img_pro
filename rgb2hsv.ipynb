```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the img
img = plt.imread('/content/face0006.jpeg')
imgplot = plt.imshow(img)
plt.show()


img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(img_hsv)

imgplot = plt.imshow(v, cmap='gray', vmin = 0, vmax = 255)
plt.show()
imgplot = plt.imshow(s, cmap='gray', vmin = 0, vmax = 255)
plt.show()
imgplot = plt.imshow(h, cmap='gray', vmin = 0, vmax = 255)
plt.show()
```
